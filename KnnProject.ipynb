```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('sonar.all-data.csv')
```

```python
df
```

```python
sns.scatterplot(data=df,x='Freq_1',y='Freq_2',hue='Label')
```

```python
df['target']= df['Label'].map({'R':0,'M':1})
```

```python
df
```

```python
df=df.drop('Label',axis=1)
```

```python
df.corr()['target'].sort_values().tail(6) # to get top 5 values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('target',axis=1)
y=df['target']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
```

```python
sc= StandardScaler()
knn = KNeighborsClassifier()
```

```python
sc_xtrain = sc.fit_transform(x_train)
sc_xtest = sc.transform(x_test)
```

```python
operation = [('sc',sc),('knn',knn)]
```

```python
pipe = Pipeline(operation)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_val = list(range(1,30))
```

```python
param_grid = {'knn__n_neighbors':k_val}
```

```python
param_grid
```

```python
full_classifier = GridSearchCV(pipe,param_grid,cv=5,scoring='accuracy')
```

```python
full_classifier.fit(x_train,y_train)
```

```python
full_classifier.best_estimator_.get_params()
```

```python
pd.DataFrame(full_classifier.cv_results_)
```

```python
pd.DataFrame(full_classifier.cv_results_)['mean_test_score'].plot()
```

```python
y_pred = full_classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_pred,y_test))
```
